```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Dataset .csv")
```

```python
df = df.dropna(subset=['Aggregate rating', 'Votes', 'Restaurant Name'])
df.reset_index(drop=True, inplace=True)
```

```python
rating_dist = df['Aggregate rating'].value_counts().sort_index()
rating_dist

rating_dist.plot(kind='bar', title='Restaurant Ratings Distribution')
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()

df['Votes'].mean()
```

```python
restaurant_counts = df['Restaurant Name'].value_counts()
chains = restaurant_counts[restaurant_counts > 1]
chains.head()

top_chains = chains.head(10)
top_chains.plot(kind='bar', title='Top 10 Restaurant Chains')
plt.ylabel("Number of Outlets")
plt.show()

chain_names = top_chains.index
chain_ratings = df[df['Restaurant Name'].isin(chain_names)] \
                    .groupby('Restaurant Name')['Aggregate rating'].mean()

chain_ratings

```

```python
top_voted = df.groupby('Restaurant Name')['Votes'].mean().sort_values(ascending=False).head(10)
top_voted

top_voted.plot(kind='bar', title='Top 10 Restaurants by Votes')
plt.ylabel("Average Votes")
plt.show()
```

## Insights & Conclusion

1. Most restaurants in the dataset have aggregate ratings between 3.0 and 4.0, indicating generally positive customer feedback.
2. The average number of votes received by restaurants is approximately 157, showing moderate customer engagement.
3. Several restaurant chains are present in the dataset, which indicates expansion of popular brands across multiple locations.
4. Restaurants with higher popularity (more outlets and votes) tend to maintain stable and competitive ratings.
